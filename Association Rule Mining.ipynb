```python
import pandas as pd

df = pd.read_csv('C:\\Users\\dhanu\\Downloads\\apriori.csv')
df=df.head(5000)
data = []
tid = 1
for index, row in df.iterrows():
    items = [str(cell) for cell in row if pd.notnull(cell)]
    data.append([str(tid), items])
    tid += 1

print(data)
```

```python
init = []
for i in data:
    for q in i[1]:
        if(q not in init):
            init.append(q)
init = sorted(init)
print(init)

# Specify the desired support value
support_threshold = 50

# Calculate the support threshold based on the desired percentage
s = support_threshold
```

```python
from collections import Counter
c = Counter()
for i in init:
    for d in data:
        if(i in d[1]):
            c[i]+=1
print("C1:")
for i in c:
    print(str([i])+": "+str(c[i]))
print()

l = Counter()
for i in c:
    if(c[i] >= s):
        l[frozenset([i])]+=c[i]
print("L1:")
for i in l:
    print(str(list(i))+": "+str(l[i]))
print()
```

```python
pl = l
pos = 1
for count in range (2,1000):
    nc = set()
    temp = list(l)
    for i in range(0,len(temp)):
        for j in range(i+1,len(temp)):
            t = temp[i].union(temp[j])
            if(len(t) == count):
                nc.add(temp[i].union(temp[j]))
    nc = list(nc)
    c = Counter()
    for i in nc:
        c[i] = 0
        for q in data:
            temp = set(q[1])
            if(i.issubset(temp)):
                c[i]+=1
    print("C"+str(count)+":")
    for i in c:
        print(str(list(i))+": "+str(c[i]))
    print()

    l = Counter()
    for i in c:
        if(c[i] >= s):
            l[i]+=c[i]
    print("L"+str(count)+":")
    for i in l:
        print(str(list(i))+": "+str(l[i]))
    print()
    if(len(l) == 0):
        break
    pl = l
    pos = count

print("Result: ")
print("L"+str(pos)+":")
for i in pl:
    print(str(list(i))+": "+str(pl[i]))
print()
```

```python
minimum_confidence = 40.0  
K = 5 

from itertools import combinations
from heapq import nlargest

top_rules = [] 

for l in pl:
    c = [frozenset(q) for q in combinations(l, len(l) - 1)]
    mmax = 0
    for a in c:
        b = l - a
        ab = l
        sab = 0
        sa = 0
        sb = 0
        for q in data:
            temp = set(q[1])
            if a.issubset(temp):
                sa += 1
            if b.issubset(temp):
                sb += 1
            if ab.issubset(temp):
                sab += 1
        confidence_a_to_b = sab / sa * 100
        confidence_b_to_a = sab / sb * 100
        if confidence_a_to_b > mmax and confidence_a_to_b >= minimum_confidence:
            mmax = confidence_a_to_b
        if confidence_b_to_a > mmax and confidence_b_to_a >= minimum_confidence:
            mmax = confidence_b_to_a
        if confidence_a_to_b >= minimum_confidence:
            top_rules.append((a, b, confidence_a_to_b))
        if confidence_b_to_a >= minimum_confidence:
            top_rules.append((b, a, confidence_b_to_a))

# Get the top-K association rules based on confidence
top_k_rules = nlargest(K, top_rules, key=lambda x: x[2])

# Display the top-K association rules
print(f"Top-{K} Association Rules:")
for i, (a, b, confidence) in enumerate(top_k_rules, start=1):
    print(f"{list(a)} -> {list(b)} (Confidence: {confidence}%)")
```
